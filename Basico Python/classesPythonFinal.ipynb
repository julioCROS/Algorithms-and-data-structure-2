```python
class Autor():      
    
    def __init__(self, nome, data_nasc):
        self.nome = nome
        self.data_nasc = data_nasc
    
    @property
    def nome(self):
        return (self.prim_nome + " " + self.segund_nome + " " + self.ultim_nome) if len(self.arr_nomes) >= 3 else (self.prim_nome + " " + self.ultim_nome)

    @nome.setter
    def nome(self, nome_complt):
        self.arr_nomes = nome_complt.split(" ")
        self.prim_nome = self.arr_nomes[0] if len(self.arr_nomes) > 0 else ""
        
        # O nome do meio é opcional
        # então caso o autor cadastrado seja: "Julio Souza"
        # "Julio" será considerado como 1o nome e "Souza" como ultimo nome, 
        # e não haverá nome do meio
        
        self.segund_nome = " ".join(self.arr_nomes[int(len(self.arr_nomes)/2) : int(len(self.arr_nomes)/2) + 1]) if len(self.arr_nomes) >= 3 else ""
        self.ultim_nome = " ".join(self.arr_nomes[len(self.arr_nomes) - 1:])  if len(self.arr_nomes) >= 3 else " ".join(self.arr_nomes[1:])          
      
        
    def nome_como_citado(self):
        self.nome_citado = self.ultim_nome.upper() + " " + self.prim_nome[0] + "."
        return self.nome_citado
    
    def __str__(self):
        return str([self.nome_como_citado()])
    
    def __repr__(self):
        return str(self)
    
```

```python
class Livro():
    
    def __init__(self, titulo, ano, autor:list):
        self.titulo = titulo
        self.ano = ano
        self.autor = autor
        
    @property
    def titulo(self):
        return self._titulo
    
    @titulo.setter
    def titulo(self, livro_titulo):
        if livro_titulo == "":
            raise ValueError("Error, titulo não pode ser vazio")
        self._titulo = livro_titulo        
    
    def __str__(self):
        return str([self.titulo, self.ano, self.autor])
    
    def __repr__(self):
        return str(self) 
        
```

```python
class Biblioteca():

    def __init__(self, lista_livros: list):
        self.lista_livros = lista_livros
        self.livros_por_autor = []

    def livros_por_autor_func(self):
        for i in range (len(self.lista_livros)):
            for j in range (len(self.lista_livros[i].autor)):
                if self.lista_livros[i].autor[j] not in self.livros_por_autor:
                    self.livros_por_autor.append((self.lista_livros[i].autor[j], self.livro_por_autor_aux(self.lista_livros[i].autor[j])))
        return self.livros_por_autor
    
    def livro_por_autor_aux(self, autor):
        livrosAutor = []
        for i in range (len(self.lista_livros)):
            for j in range (len(self.lista_livros[i].autor)):
                if autor in self.lista_livros[i].autor:
                    if self.lista_livros[i].titulo not in livrosAutor:
                        livrosAutor.append(self.lista_livros[i].titulo) 
        return livrosAutor
    
    def __str__(self):
        return str(self.lista_livros)
    
    def __repr__(self):
        return str(self)
        
```
